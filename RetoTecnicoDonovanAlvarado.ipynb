```python
import pandas as pd
import urllib3 as ur
```

```python
#Getting the dataframe and connection
http = ur.PoolManager()
r = http.request('GET', 'https://api.stackexchange.com/2.2/search?order=desc&sort=activity&intitle=perl&site=stackoverflow')
df = pd.read_json(r.data)
items = pd.json_normalize(df['items'])
```

```python
#Format Dates
items['creation_date'] = pd.to_datetime(items['creation_date'],unit='s')
items['last_activity_date'] = pd.to_datetime(items['last_activity_date'],unit='s')
items['last_edit_date'] = pd.to_datetime(items['last_edit_date'],unit='s')
items['closed_date'] = pd.to_datetime(items['closed_date'],unit='s')
items.head()
```

```python
#Getting the questions answered and not answered
a = len(items[items['is_answered'] == True])
b = len(items[items['is_answered'] == False])
print("Se contestaron: " +str(a)+ " Faltaron:"+str(b))
```

```python
#Getting answered with max owners
max_owners = items[items['owner.reputation']==items['owner.reputation'].max()]
max_owners
```

```python
new = items[items['creation_date']==items['creation_date'].max()]
new
```
